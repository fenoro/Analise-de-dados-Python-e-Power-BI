```python
# Importando Bibliotecas
import pandas as pd
```

```python
# Importando base de dados
amostra_360 = pd.read_excel("amostra_360.xlsx")
```

```python
# primeiras views
print(amostra_360.info())
print()
print(amostra_360.shape)
print()
print(amostra_360.head(3))

# Selecionar a coluna de chave para separar em 9 tabelas: hc, hp, obra, retirada, bloqueio, desbloqueio, bairro, municipio e estado
print(amostra_360.nunique().idxmax())
print(amostra_360["id_hc"].nunique())
        # portando chave: id_hc
```

```python
# Criação DF homes_passed e exclusão de linhas e colunas vazias

homes_passed = amostra_360[["id_hc", "no_viabilidade_hp", "ds_estacao_abastecedora_hp", "cd_tipo_survey_hp","hc_max","ds_cep_hp",
                            "ds_cod_localidade_hp", "ds_uf_hp","cd_latitude_hp", "cd_longitude_hp", "diretoria", "cd_tipo_rede_hp",
                            "nu_ucs_residenciais_hp", "nu_ucs_comerciais_hp", "dt_lcto_comercial", "status_hp"]].copy()

print(homes_passed.shape)
print(homes_passed.isnull().sum())
homes_passed.dropna(subset=["no_viabilidade_hp", "ds_estacao_abastecedora_hp", "cd_tipo_survey_hp","hc_max","ds_cep_hp",
                            "ds_cod_localidade_hp", "ds_uf_hp","cd_latitude_hp", "cd_longitude_hp", "diretoria", "cd_tipo_rede_hp",
                            "nu_ucs_residenciais_hp", "nu_ucs_comerciais_hp", "dt_lcto_comercial", "status_hp"], 
                    how="all", axis=0, inplace=True)
print(homes_passed.isnull().sum())
homes_passed.dropna(how="all", axis=1, inplace=True)
print(homes_passed.isnull().sum())
print(homes_passed.shape)
```

```python
# Checagem de tipos homes_passed
print(homes_passed.select_dtypes("datetime").columns)
print(homes_passed.select_dtypes("number").columns)
```

```python
# Criação DF homes_connected e exclusão de linhas e colunas vazias

homes_connected = amostra_360[["id_hc", "status_hc", "status_gpon", "status_iptv", "status_voip", "status_banda_larga", "velocidade_hc", 
                               "ds_uf_hc","cd_ibge_municipio", "cd_cep_hc", "regiao", "sistema_origem_os", "segmento_hc", "ont_fabricante",
                               "tipo_infra", "prestadora_servico", "produto", "dt_solicitacao", "ic_prazo_instalacao_os", "ic_cancelado",
                               "status_ordem_instalacao", "ds_uf_os", "ds_municipio_os", "ds_estado_instalacao"]].copy()

print(homes_connected.shape)
print(homes_connected.isnull().sum())
homes_connected.dropna(subset=["status_hc", "status_gpon", "status_iptv", "status_voip", "status_banda_larga", "velocidade_hc", 
                               "ds_uf_hc", "cd_ibge_municipio", "cd_cep_hc", "regiao", "sistema_origem_os", "segmento_hc", "ont_fabricante",
                               "tipo_infra", "prestadora_servico", "produto", "dt_solicitacao", "ic_prazo_instalacao_os", "ic_cancelado",
                               "status_ordem_instalacao", "ds_uf_os", "ds_municipio_os", "ds_estado_instalacao"], 
                    how="all", axis=0, inplace=True)
print(homes_connected.isnull().sum())
homes_connected.dropna(how="all", axis=1, inplace=True)
print(homes_connected.isnull().sum())
print(homes_connected.shape)
```

```python
# Checagem de tipos homes_connected
print(homes_connected.select_dtypes("datetime").columns)
```

```python
# Criação DF obra e exclusão de linhas e colunas vazias

homes_connected_obra = amostra_360[["id_hc", "status_ordem_obra", "f11_", "tipo_pendencia_obra", "status_atividade_obra", "f21_"]].copy()
homes_connected_obra = homes_connected_obra.rename(columns={"f11_":"dt_ordem_obra", "f21_":"dt_atividade_obra"})

print(homes_connected_obra.shape)
print(homes_connected_obra.isnull().sum())
homes_connected_obra.dropna(subset=["status_ordem_obra", "dt_ordem_obra", "tipo_pendencia_obra", "status_atividade_obra", "dt_atividade_obra"], 
                    how="all", axis=0, inplace=True)
print(homes_connected_obra.isnull().sum())
homes_connected_obra.dropna(how="all", axis=1, inplace=True)
print(homes_connected_obra.isnull().sum())
print(homes_connected_obra.shape)

# Criação DF retirada e exclusão de linhas e colunas vazias
homes_connected_retirada = amostra_360[["id_hc", "status_ordem_retirada", "f23_", "tipo_pendencia_retirada", "ds_estado_retirada"]].copy()
homes_connected_retirada = homes_connected_retirada.rename(columns={"f23_":"dt_ordem_retirada"})

print(homes_connected_retirada.shape)
print(homes_connected_retirada.isnull().sum())
homes_connected_retirada.dropna(subset=["status_ordem_retirada", "dt_ordem_retirada", "tipo_pendencia_retirada", "ds_estado_retirada"], 
                    how="all", axis=0, inplace=True)
print(homes_connected_retirada.isnull().sum())
homes_connected_retirada.dropna(how="all", axis=1, inplace=True)
print(homes_connected_retirada.isnull().sum())
print(homes_connected_retirada.shape)

# Criação DF bloqueio e exclusão de linhas e colunas vazias
homes_connected_bloqueio = amostra_360[["id_hc", "status_ordem_bloqueio", "f43_", "tipo_pendencia_bloqueio", "ds_estado_bloqueio"]].copy()
homes_connected_bloqueio = homes_connected_bloqueio.rename(columns={"f43_":"dt_ordem_bloqueio"})

print(homes_connected_bloqueio.shape)
print(homes_connected_bloqueio.isnull().sum())
homes_connected_bloqueio.dropna(subset=["status_ordem_bloqueio", "dt_ordem_bloqueio", "tipo_pendencia_bloqueio", "ds_estado_bloqueio"], 
                    how="all", axis=0, inplace=True)
print(homes_connected_bloqueio.isnull().sum())
homes_connected_bloqueio.dropna(how="all", axis=1, inplace=True)
print(homes_connected_bloqueio.isnull().sum())
print(homes_connected_bloqueio.shape)

# Criação DF desbloqueio e exclusão de linhas e colunas vazias
homes_connected_desbloqueio = amostra_360[["id_hc", "status_ordem_desbloqueio", "f53_", "tipo_pendencia_desbloqueio", "ds_estado_desbloqueio"]].copy()
homes_connected_desbloqueio = homes_connected_desbloqueio.rename(columns={"f53_":"dt_ordem_desbloqueio"})
print(homes_connected_desbloqueio.shape)
print(homes_connected_desbloqueio.isnull().sum())
homes_connected_desbloqueio.dropna(subset=["status_ordem_desbloqueio", "dt_ordem_desbloqueio", "tipo_pendencia_desbloqueio", "ds_estado_desbloqueio"], 
                    how="all", axis=0, inplace=True)
print(homes_connected_desbloqueio.isnull().sum())
homes_connected_desbloqueio.dropna(how="all", axis=1, inplace=True)
print(homes_connected_desbloqueio.isnull().sum())
print(homes_connected_desbloqueio.shape)
```

```python
# Checagem de tipos obra, retirada, bloqueio e desbloqueio
print(homes_connected_obra.select_dtypes("datetime64[ns, UTC]").columns)
print(homes_connected_retirada.select_dtypes("datetime64[ns, UTC]").columns)
print(homes_connected_bloqueio.select_dtypes("datetime64[ns, UTC]").columns)
print(homes_connected_desbloqueio.select_dtypes("datetime64[ns, UTC]").columns)

homes_connected_obra["dt_ordem_obra"] = pd.to_datetime(homes_connected_obra["dt_ordem_obra"])
homes_connected_obra["dt_atividade_obra"] = pd.to_datetime(homes_connected_obra["dt_atividade_obra"], format='mixed')
homes_connected_retirada["dt_ordem_retirada"] = pd.to_datetime(homes_connected_retirada["dt_ordem_retirada"], format='mixed')
homes_connected_bloqueio["dt_ordem_bloqueio"] = pd.to_datetime(homes_connected_bloqueio["dt_ordem_bloqueio"])
homes_connected_desbloqueio["dt_ordem_desbloqueio"] = pd.to_datetime(homes_connected_desbloqueio["dt_ordem_desbloqueio"])

print(homes_connected_obra.select_dtypes("datetime64[ns, UTC]").columns)
print(homes_connected_obra.select_dtypes("datetime64[ns]").columns)
print(homes_connected_retirada.select_dtypes("datetime64[ns, UTC]").columns)
print(homes_connected_bloqueio.select_dtypes("datetime64[ns, UTC]").columns)
print(homes_connected_desbloqueio.select_dtypes("datetime64[ns, UTC]").columns)
```

```python
# Criação DF localidades
    # cep como código de bairro
    # localidade_hp como código de municipio_hp
    # ibge_municipio como código de municipio_hc

localidades_hp = amostra_360[["ds_cep_hp", "ds_bairro_hp", "ds_cod_localidade_hp", "ds_municipio_hp", "ds_uf_hp"]].copy()
localidades_hp = localidades_hp.rename(columns={"ds_cep_hp":"cd_bairro", "ds_bairro_hp":"ds_bairro", "ds_cod_localidade_hp":"cd_municipio",
                                                "ds_municipio_hp":"ds_municipio", "ds_uf_hp":"sg_estado"})

localidades_hc = amostra_360[["cd_cep_hc", "ds_bairro_hc", "cd_ibge_municipio", "ds_municipio_hc", "ds_uf_hc"]].copy()
localidades_hc = localidades_hc.rename(columns={"cd_cep_hc":"cd_bairro", "ds_bairro_hc":"ds_bairro", "cd_ibge_municipio":"cd_municipio", 
                                                "ds_municipio_hc":"ds_municipio", "ds_uf_hc":"sg_estado"})
print(localidades_hp.shape, localidades_hc.shape)

localidades = pd.concat([localidades_hp, localidades_hc])

print(localidades.shape)
print(localidades.head(3))

print(localidades.isnull().sum())
localidades.dropna(how="all", axis=0, inplace=True)

print(localidades.isnull().sum())
localidades.dropna(how="all", axis=1, inplace=True)

print(localidades.isnull().sum())
print(localidades.shape)

localidades = localidades.drop_duplicates(subset=["cd_bairro","ds_bairro","cd_municipio", "ds_municipio","sg_estado"])
print(localidades.isnull().sum())
print(localidades.shape)
```

```python
localidades_bairro = localidades[["cd_bairro","ds_bairro","cd_municipio","sg_estado"]].copy()

print(localidades_bairro.shape)

print(localidades_bairro.isnull().sum())
localidades_bairro.dropna(how="all", axis=0, inplace=True)

print(localidades_bairro.isnull().sum())
localidades_bairro.dropna(how="all", axis=1, inplace=True)

print(localidades_bairro.isnull().sum())
print(localidades_bairro.shape)

localidades_bairro = localidades_bairro.drop_duplicates(subset=["cd_bairro","ds_bairro","cd_municipio","sg_estado"])
print(localidades_bairro.isnull().sum())
print(localidades_bairro.shape)


localidades_municipio = localidades[["cd_municipio", "ds_municipio","sg_estado"]].copy()

print(localidades_municipio.shape)

print(localidades_municipio.isnull().sum())
localidades_municipio.dropna(how="all", axis=0, inplace=True)

print(localidades_municipio.isnull().sum())
localidades_municipio.dropna(how="all", axis=1, inplace=True)

print(localidades_municipio.isnull().sum())
print(localidades_municipio.shape)

localidades_municipio = localidades_municipio.drop_duplicates(subset=["cd_municipio", "ds_municipio","sg_estado"])
print(localidades_municipio.isnull().sum())
print(localidades_municipio.shape)

localidades_estado = localidades[["sg_estado"]].copy()
print(localidades_estado.shape)

print(localidades_estado.isnull().sum())
localidades_estado.dropna(how="all", axis=0, inplace=True)

print(localidades_estado.isnull().sum())
localidades_estado.dropna(how="all", axis=1, inplace=True)

print(localidades_estado.isnull().sum())
print(localidades_estado.shape)

localidades_estado = localidades_estado.drop_duplicates(subset=["sg_estado"])
print(localidades_estado.isnull().sum())
print(localidades_estado.shape)
```

```python
homes_passed.to_csv("homes_passed.csv", index=False)
homes_connected.to_csv("homes_connected.csv", index=False)
homes_connected_bloqueio.to_csv("homes_connected_bloqueio.csv", index=False)
homes_connected_desbloqueio.to_csv("homes_connected_desbloqueio.csv", index=False)
homes_connected_obra.to_csv("homes_connected_obra.csv", index=False)
homes_connected_retirada.to_csv("homes_connected_retirada.csv", index=False)
localidades_bairro.to_csv("localidades_bairro.csv", index=False)
localidades_municipio.to_csv("localidades_municipio.csv", index=False)
localidades_estado.to_csv("localidades_estado.csv", index=False)
```
